Question 2 — Extracting Tesla Revenue Data (1 point)

```python
import pandas as pd

url_tesla = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Read tables directly
tables = pd.read_html(url_tesla)

# Tesla revenue table is at index 1
tesla_revenue = tables[1]
tesla_revenue.columns = ["Date", "Revenue"]

# Clean the Revenue column
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].replace({'\$': '', ',': ''}, regex=True)
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
tesla_revenue.reset_index(drop=True, inplace=True)

print(tesla_revenue.head())
```

Question 4 — Extracting GameStop Revenue Data (1 point)

```python
url_gme = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Read tables directly
gme_tables = pd.read_html(url_gme)

# GameStop revenue table is at index 0
gme_revenue = gme_tables[0]
gme_revenue.columns = ["Date", "Revenue"]

# Clean the Revenue column
gme_revenue["Revenue"] = gme_revenue["Revenue"].replace({'\$': '', ',': ''}, regex=True)
gme_revenue = gme_revenue[gme_revenue["Revenue"] != ""]
gme_revenue.reset_index(drop=True, inplace=True)

print(gme_revenue.head())
```

Question 5 — Tesla Stock and Revenue Dashboard (2 points)

```python
import yfinance as yf

# Create ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Get full historical stock data
tesla_data = tesla.history(period="max")

# Reset index to make Date a column
tesla_data.reset_index(inplace=True)

# Check first 5 rows
print(tesla_data.head())

```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10,5))

# Plot Tesla stock price
ax1.plot(tesla_data['Date'], tesla_data['Close'], color='tab:blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Tesla Stock Price', color='tab:blue')

# Plot Tesla revenue (make sure tesla_revenue exists too)
ax2 = ax1.twinx()
ax2.plot(tesla_revenue['Date'], tesla_revenue['Revenue'], color='tab:red')
ax2.set_ylabel('Revenue (USD)', color='tab:red')

plt.title('Tesla Stock Price vs Quarterly Revenue')
plt.show()
```
